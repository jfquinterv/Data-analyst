The portfolio return $R_p$ is the weighted sum of the individual asset returns:

$$
R_p = \sum_{i=1}^{k} w_i R_i
$$

The portfolio variance $\sigma^2_p$ can be expressed in terms of individual asset standard deviations $\sigma_i$, $\sigma_j$, and the correlation coefficient $\rho_{ij}$ between the returns of assets $i$ and $j$:

$$
\sigma^2_p = \sum_{i=1}^{k} \sum_{j=1}^{k} w_i w_j \sigma_i \sigma_j \rho_{ij}
$$

The portfolio variance is often simplified using the covariance $\sigma_{ij}$ between the returns of assets $i$ and $j$:

$$
\sigma^2_p = \sum_{i=1}^{k} \sum_{j=1}^{k} w_i w_j \sigma_{ij}
$$

Here, $\sigma_{ij}$ is equivalent to $\sigma_i \sigma_j \rho_{ij}$, and it represents the covariance between asset $i$ and asset $j$.


volatility depends on the covariance

We start by constructing the weight vector $ \mathbf{w} $, which is a $ k \times 1 $ vector of weights, and $ \mathbf{R} $, which is the $ k \times 1 $ vector of asset returns.

The portfolio return $ R_p $ is given by the formula:

$$
R_p = \mathbf{w}^T\mathbf{R}
$$

The weight vector $ \mathbf{w} $ is defined as:

$$
\mathbf{w} = \begin{bmatrix}
w_1 \\
w_2 \\
\vdots \\
w_k
\end{bmatrix}
$$

And the return vector $ \mathbf{R} $ is:

$$
\mathbf{R} = \begin{bmatrix}
R_1 \\
R_2 \\
\vdots \\
R_k
\end{bmatrix}
$$


Given the weight vector $ \mathbf{w} $, the covariance matrix $ \Sigma $ is a $ k \times k $ matrix representing the covariances between the asset returns. It is formally defined as:

$$
\Sigma = \begin{bmatrix}
\sigma_{11} & \sigma_{12} & \cdots & \sigma_{1k} \\
\sigma_{21} & \sigma_{22} & \cdots & \sigma_{2k} \\
\vdots & \vdots & \ddots & \vdots \\
\sigma_{k1} & \sigma_{k2} & \cdots & \sigma_{kk}
\end{bmatrix}
$$

Where $ \sigma_{ij} $ represents the covariance between the returns of asset $ i $ and asset $ j $.

The variance of the portfolio $ \sigma_p^2 $ is then given by the formula:

$$
\sigma_p^2 = \mathbf{w}^T \Sigma \mathbf{w}
$$


fist you start choosing one portafolios that lies on the efficient frontier curve, you focus on the lowest return portafolio and the highest return maybe they are not in the efficient frontier, but they are in the region of posible portafolios, the case you put all your money in the lowest return portafolio and the case when you put all you'r money in the highest return portafolio, then you start building portafolios that esentially lies between this portafolios and every single time we run the optimizer to fin the minimum posible variance portafolio. you split the space between the highest and lowest return portafolio, you split it into a grid, in every grid you run the optimizer


The quadratic optimization problem depicted aims to determine the optimal asset weights to minimize the risk of a portfolio. The problem is formalized as follows:

Minimize the portfolio variance:
$$
\min \frac{1}{2} \mathbf{w}^T \Sigma \mathbf{w}
$$

Subject to the constraints that ensure the portfolio has a specified expected return, the weights sum up to 1 (full investment of the capital), and no short selling is allowed (weights are non-negative):
$$
\mathbf{w}^T \mathbf{R} = r_0
$$
$$
\mathbf{w}^T \mathbf{1} = 1
$$
$$
\mathbf{w} \geq 0
$$

The optimization is typically solved using quadratic programming methods.

**how do i transform two case assets efficient frontier to a N assets case?**
`The key issue is the weight vector`

we find the weigh vector with an optimization problem when we minimize the vol given a return target. For that we use the quadratic equation  and use the library scipy

we know the largest and lowest return, we're going to pick some number in between for example 15%, the question is `how can i figure out the weights that gives me the lowest possible volatility for that 15% target return`, then i do it for a whole range of target returns that gives me every single weight vector along the way


````python
from scipy.optimize import minimize
import numpy as np

def minimize_vol(target, er, cov):
    """
    Perform portfolio optimization by finding the weights that minimize volatility.

    Parameters:
    - target: The desired return for the portfolio.
    - er: A numpy array containing the expected returns of each asset.
    - cov: The covariance matrix representing the relationships between asset returns.

    Returns:
    - Optimized portfolio weights that meet the specified target return and constraints.
    """

    # Number of assets in the portfolio
    n = er.shape[0]

    # Initial guess: Equal weight for each asset (1/n)
    initial_guess = np.repeat(1/n, n)

    # Define the bounds for each weight (between 0 and 1)
    # This creates a tuple of (0.0, 1.0) pairs, one for each asset
    bounds = ((0.0, 1.0),) * n

    # Constraint 1: Portfolio return should meet the target return
    # 'type' specifies the type of constraint ('eq' for equality)
    # 'args' passes additional required arguments to the function
    # The function calculates the difference between target return and actual return
    # If the constraint is met, this function will return zero
    return_is_target = {
        'type': 'eq',
        'args': (er,), #one element tuple
        'fun': lambda weights, er: target - r.portafolio_r(er, weights)
    }

    #lambda functions is used when we want to create an annonymus function, without formaly create the function with keyword def, lambda function can have any arguments but only one expresion, it its syntax it isn't formaled
    #has curly brakets, the curly brakets in python are for python dictionaries.
    
    
    # Constraint 2: The sum of weights must be equal to 1
    # This ensures that the total allocated percentage is 100%
    weights_sum_1 = {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights) - 1
    }

    # Optimization function
    # 'minimize' function is used for finding the minimum of a function
    # 'r.portafolio_vol()' is the function being minimized (portfolio volatility)
    # 'method' specifies the algorithm used for optimization ('SLSQP' for sequential least squares programming)
    # 'options' are additional settings for the optimizer ('disp': False suppresses output)
    # 'constraints' are the defined constraints that the solution must satisfy
    results = minimize(
        lambda weights: r.portafolio_vol(weights, cov),  # Modified to include the required arguments
        initial_guess,
        method="SLSQP",
        options={'disp': False},
        constraints=(return_is_target, weights_sum_1),
        bounds=bounds
    )

    # Return the optimized weights
    return results.x



def optimal_weights(er,cov, n_points):
    """
    list of weights to run the optimizer on to minimize the vol
    """
    target_rs=np.linspace(er.min(),er.max(),n_points)
    weights=[minimize_vol(target_returns, er, cov) for target_returns in target_rs]
    return weights
    
def plot_ef(er,cov, n_points):
    """
    efficient frontier
    """
    weights=optimal_weights(er,cov, n_points)
    rts=[r.portafolio_r(er,w) for w in weights]
    vols=[r.portafolio_vol(w,cov)for w in weights]
    ef=pd.DataFrame({"Returns":rts,
                     "Volatility":vols})
    return ef.plot.line(x="Volatility", y="Returns", style='.-')
````

````python
import risk as r
from scipy.optimize import minimize
def msr(risk_free, er, cov):
    """
    Risk_free_rate+ER+COV->W
    """

    # Number of assets in the portfolio
    n = er.shape[0]

    # Initial guess: Equal weight for each asset (1/n)
    initial_guess = np.repeat(1/n, n)

    # Define the bounds for each weight (between 0 and 1)
    # This creates a tuple of (0.0, 1.0) pairs, one for each asset
    bounds = ((0.0, 1.0),) * n

    # Constraint 1: The sum of weights must be equal to 1
    # This ensures that the total allocated percentage is 100%
    weights_sum_1 = {
        'type': 'eq',
        'fun': lambda weights: np.sum(weights) - 1
    }

    # Optimization function
    # 'options' are additional settings for the optimizer ('disp': False suppresses output)
    # we can embed a function within a function
    # the objetive function is the sharpe ratio, we can maximize using a minimization, minimizing
    # the negative of the objective funtion. ¿how do i define the negative of sharpe ratio?
   
    def neg_sharpe_ratio(weights, risk_free, er, cov):
        """
        return negative sharpe ratio given the weights
        """
        rt=r.portafolio_r(er,weights)
        vol=r.portafolio_vol(weights, cov)
        return -(rt-risk_free)/vol
        
    results = minimize(neg_sharpe_ratio, initial_guess,
                   args=(risk_free, er, cov), method="SLSQP",
                   options={'disp': False},
                   constraints=(weights_sum_1),
                   bounds=bounds)
    # Return the optimized weights
    return results.x
````

````python
ax=r.plot_ef(er,cov,30)
ax.set_xlim(left=0)
rf=0.1
w_msr=msr(rf,er,cov)
rt_msr=r.portafolio_r(er,w_msr)
vol_msr=r.portafolio_vol(w_msr,cov)
cml_x=[0,vol_msr]
cml_y=[rf,rt_msr]
ax.plot(cml_x,cml_y,color="green",linestyle="dashed")
````
